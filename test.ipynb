```python
import matplotlib.pyplot as plt

# Define models
models = ['Unet', 'ResUnet', 'EffUnet', 'FPNUnet', 'UNet++', 'Unet++ with Deep Supervision']

# Define metrics
train_loss = [0.0106, 0.0112, 0.1821, 0.1813, 0.1829, 0.2322]
val_loss = [0.0237, 0.01693, 0.2232, 0.2267, 0.2415, 0.3259]
train_dice = [0.669, 0.714, 0.7354, 0.7853, 0.7239, 0.7339]
val_dice = [0.596, 0.623, 0.658, 0.6626, 0.6636, 0.6773]
accuracy = [0.8772, 0.8948, 0.9046, 0.8934, 0.8927, 0.9147]

# Plotting
plt.figure(figsize=(14, 7))

plt.plot(models, train_loss, marker='o', label='Train Loss')
plt.plot(models, val_loss, marker='o', label='Val Loss')
plt.plot(models, train_dice, marker='o', label='Train Dice')
plt.plot(models, val_dice, marker='o', label='Val Dice')
plt.plot(models, accuracy, marker='o', label='Accuracy')

plt.title('Model Comparison for Pneumothorax Segmentation')
plt.xlabel('Model')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
